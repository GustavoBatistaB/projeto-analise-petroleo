Sobre as empresas

```python
# AS 10 Empresas Com um maior numero de Ocorrencia
rank_ocorrencia = df.groupby('Empresa')['Empresa'].count()
print('Rank das 10 Empreas com mais ocorrencias ')
rank_ocorrencia.sort_values(ascending=False).head(10)
```

```python

print('AS 10 Empresas Com um maior numero de Feridos Graves')
rank_feridos_graves = df.groupby('Empresa')['Numero_de_feridos_graves'].sum()
rank_feridos_graves.sort_values(ascending=False).head(10)
```

```python

print('Media das Empresas -  Feridos Graves')
rank_media_feridos_graves = df.groupby('Empresa')['Numero_de_feridos_graves'].mean()
rank_media_feridos_graves.sort_values(ascending=False).head(10)
```

```python
print('AS 10 Empresas Com um maior numero de Fatalidades')
rank_fatalidade = df.groupby('Empresa')['Numero_de_fatalidades'].sum()
rank_fatalidade.sort_values(ascending=False).head(10)
```

```python
print('AS 69 Empresas Com um Menor numero de Fatalidades')
rank_menor_fatalidade = df.groupby('Empresa')['Numero_de_fatalidades'].sum()
rank_menor_fatalidade.sort_values(ascending=True).head(69)
```

Sobre vítimas

```python
print(f' Total de Pessoas que o feriento foi grave {df['Numero_de_feridos_graves'].sum()}')
print(f' Total de Pessoas que o insidente foi fatal  {df['Numero_de_fatalidades'].sum()}')
```

```python
print('Insidente Com o maior numero de Fatalidade')
df[df['Numero_de_fatalidades']== df['Numero_de_fatalidades'].max()]
```

```python
(f'Total de ocorrencias que não registraram Feridos e fatalidades {len(df[(df["Numero_de_feridos_graves"] == 0) & (df["Numero_de_fatalidades"] == 0)])}')
```

```python
correlacao = df['Numero_de_feridos_graves'].corr(df['Numero_de_fatalidades'])
(f'A Correlação entre as Colunas Numero_de_feridos_graves e Numero_de_fatalidades é: {correlacao:.2f}')
```


Sobre instalações

```python
print('As 10 Empresas Com mais Instalações ')
rank_empresa_instalacao = df.groupby('Empresa')['Instalacao'].nunique()
rank_empresa_instalacao.sort_values(ascending=False).head(10)
```

```python
print('Rank Instalações Com mais acidentes Fatais ')
rank_instalacao_acidente_fatal = df.groupby('Instalacao')['Numero_de_fatalidades'].sum()
rank_instalacao_acidente_fatal.sort_values(ascending=False).head(10)
```

```python
print('Rank das 10 instalaçoes com mais ocorrencias ')
rank_instalacao_ocorrencia = df.groupby('Instalacao')['Instalacao'].count()
rank_instalacao_ocorrencia.sort_values(ascending=False).head(10)
```

análise estatística

```python
print('Feridos Graves')
print(f'A Media dos feridos graves é {df['Numero_de_feridos_graves'].mean():.5f}')
print(f'A Mediana dos feridos graves é {df['Numero_de_feridos_graves'].median():.5f}')
print(f'A Desvio Padrão dos feridos graves é {df['Numero_de_feridos_graves'].std():.5f}')
print(f'A Soma dos feridos graves é {df['Numero_de_feridos_graves'].sum()}')
```

```python
print('Fatalidade')
print(f'A Media dos fatalidade é {df['Numero_de_fatalidades'].mean():.5f}')
print(f'A Mediana dos fatalidade é {df['Numero_de_fatalidades'].median():.5f}')
print(f'A Desvio Padrão dos fatalidade é {df['Numero_de_fatalidades'].std():.5f}')
print(f'A Soma das fatalidades é {df['Numero_de_fatalidades'].sum()}')
```

```python
df['Data_de_criacao'] = pd.to_datetime(df['Data_de_criacao'], format='%d-%m-%Y')
print('     Registro')
print(f'Primeiro {str(df['Data_de_criacao'].min())[0:11]}')
print(f'Ultimo {str(df['Data_de_criacao'].max())[0:11]}')
df['Ano'] = df['Data_de_criacao'].dt.year
print(f'Periodo: {df['Ano'].unique()}')
```

```python
por_ano_feridos = df.groupby('Ano')['Numero_de_feridos_graves'].sum()
print('Quantidade de Feridos Graves por Ano')
por_ano_feridos
```

```python
fatalidade_por_ano = df.groupby('Ano')['Numero_de_fatalidades'].sum()
print('Quantidade de Fatalidade por Ano')
fatalidade_por_ano
```

```python
# Ocorrencias Por Ano
ocorrencia_por_ano = df.groupby('Ano')['Numero_de_feridos_graves'].count()
print('Quantidade de Ocorrencias por Ano')
ocorrencia_por_ano
```

```python
plt.bar(fatalidade_por_ano.index, fatalidade_por_ano.values, color='brown')
plt.xlabel('Ano')
plt.ylabel('Fatalidade')
plt.title('Fatalidade por Ano')
plt.show()
```

```python
plt.bar(por_ano_feridos.index, por_ano_feridos.values, color='#1976d2')
plt.xlabel('Ano')
plt.ylabel('Feridos')
plt.title('Feridos Por Ano')
plt.show()
```

```python
plt.plot(por_ano_feridos.index, por_ano_feridos.values, marker='o', label='Feridos Graves', color='#1976d2')
plt.plot(fatalidade_por_ano.index, fatalidade_por_ano.values, marker='o', label='Fatalidades', color='brown')
plt.xlabel('Ano')
plt.ylabel('Feridos')
plt.grid(True)
plt.legend()
plt.title('Feridos x Fatalidades Por Ano')
plt.show()
```

```python

plt.bar(por_ano_feridos.index, por_ano_feridos.values, label='Feridos', color='#1976d2')
plt.bar(fatalidade_por_ano.index, fatalidade_por_ano.values, label='Fatalidades', color='brown')
plt.title('Feridos x Fatalidades por Ano')
plt.ylabel('Pessoas')
plt.legend()
plt.show()
```

```python
plt.plot(ocorrencia_por_ano.index, ocorrencia_por_ano.values, marker='o', label='Ocorrencias', color='green')
plt.ylabel('Ocorrencias')
plt.grid(True)
plt.legend()
plt.title('Ocorrencias por Ano')
plt.show()
```

```python
plt.boxplot([df['Numero_de_fatalidades'], df['Numero_de_feridos_graves']])
plt.xticks([1,2], ['Fatalidade', 'Feridos'])
plt.title('Fatalidade x Feridos')
plt.grid(True)
plt.show()
```
